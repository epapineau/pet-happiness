```python
import pandas as pd
from sqlalchemy import create_engine
import os
```

```python
petPopData = pd.read_csv("./data/pet-populations.csv")
worldHappiness = pd.read_csv("./data/world_happiness_2017.csv")
worldHappiness = worldHappiness[["Country", "Happiness.Rank", "Happiness.Score", "Economy..GDP.per.Capita.", "Health..Life.Expectancy.", "Generosity"]]
worldHappiness = worldHappiness.rename(columns = {
    "Happiness.Rank": "Happiness_Rank",
    "Happiness.Score": "Happiness_Score",
    "Economy..GDP.per.Capita.": "GDP_per_capita",
    "Health..Life.Expectancy.": "Life_expectancy"
})
```

```python
# what countries does pet dataset have in common
    # Dog dataset only presented for 19 countries
petPopData.groupby(["Country"]).count()
```

```python
# Create sqllite file
engine = create_engine('sqlite:///./db/database.sqlite')
```

```python
# Add pet population to sqllite db
petPopData.to_sql('Pet_Population_Data', con = engine)
engine.execute("SELECT * FROM Pet_Population_Data").fetchall()
```

```python
# Add world happiness to sqllite db
worldHappiness.to_sql('World_Happiness', con = engine)
engine.execute("SELECT * FROM World_Happiness").fetchall()
```
