```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
petgdp_2017 = pd.read_csv('../data/happiness_and_gdp_and_urban.csv')
petgdp_2017.head()
```

```python
pet_population = pd.read_csv('../data/pet-populations.csv')
pet_population .head()
```

```python
pet_gdp = pd.merge(petgdp_2017, pet_population)
pet_gdp.head()
```

```python
petgdp_df = pet_gdp[["Country","GDP","Pet","Population"]]
petgdp_df.head()
```

```python
petgdp_sorted = petgdp_df.sort_values(["Population"], ascending=False)
petgdp_sorted.head()
```

```python
dog_data = petgdp_sorted.query('Pet == "Dog"')
dog_data.head()
```

```python
cat_data = petgdp_sorted.query('Pet == "Cat"')
cat_data.head()
```

```python
fish_data = petgdp_sorted.query('Pet == "Fish"')
fish_data.head()
```

```python
bird_data = petgdp_sorted.query('Pet == "Bird"')
bird_data.head()
```
